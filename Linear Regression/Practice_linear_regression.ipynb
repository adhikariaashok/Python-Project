```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.cross_validation import train_test_split
```

```python
df=pd.read_csv('USA_Housing.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns
```

```python
sns.pairplot(df)
```

```python
sns.jointplot(x='Price',y='Avg. Area Number of Rooms',data=df,kind='reg')
```

```python
sns.heatmap(df.corr())
```

```python
df.columns
```

```python
x=df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms',
       'Avg. Area Number of Bedrooms', 'Area Population']]
```

```python
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=101)


```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
print(lm.intercept_)
```

```python
x_train.columns
```

```python
pd.DataFrame(lm.coef_,x.columns,columns=['Coeff'])
```

```python
predictions=lm.predict(x_test)
```

```python
predictions
```

```python
plt.scatter(predictions,y_test)
```

```python
sns.jointplot(predictions,y_test,kind='reg')
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions)
```

```python
metrics.mean_squared_error(y_test,predictions)
```

```python
customers.columns
```

```python
x=customers[[ 'Avg. Session Length', 'Time on App',
       'Time on Website', 'Length of Membership']]
```

```python
y=customers['Yearly Amount Spent']
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=101)

```

```python
from sklearn.linear_model import LinearRegression 
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
regression.coef_
```

```python

coeff=pd.DataFrame(regression.coef_,x.columns,columns=['Coeff'])

coeff
```

```python
prediction=regression.predict(x_test)
```

```python
plt.scatter(prediction,y_test)
```

```python
sns.jointplot(y_test,prediction,kind='scatter')
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,prediction)
```

```python
metrics.mean_squared_error(y_test,prediction)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,prediction))
```

```python
sns.distplot((y_test-prediction),bins=50)
```
